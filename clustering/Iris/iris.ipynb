```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
plt.scatter(X[y_hc == 0]['sepal length (cm)'], X[y_hc == 0]['sepal width (cm)'], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1]['sepal length (cm)'], X[y_hc == 1]['sepal width (cm)'], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Clusters of customers')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
plt.show()
```

```python
plt.scatter(X[y_hc == 0]['petal length (cm)'], X[y_hc == 0]['petal width (cm)'], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1]['petal length (cm)'], X[y_hc == 1]['petal width (cm)'], s = 100, c = 'blue', label = 'Cluster 2')
plt.title('Clusters of customers')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend()
plt.show()
```

```python
import seaborn as sns

X['cluster'] = y_hc
sns.pairplot(X, hue='cluster', vars=iris.feature_names)
plt.show()
```
