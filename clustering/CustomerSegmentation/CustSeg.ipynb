```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Customer.csv')
X = df.iloc[:, [1, 2]].values
X
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
X_sc = df[['Annual_Income', 'Spending_Score']].values
```

```python
plt.scatter(X_sc[y_kmeans == 0, 0], X_sc[y_kmeans == 0, 1], s = 70, c = 'red', label = 'Cluster 1')
plt.scatter(X_sc[y_kmeans == 1, 0], X_sc[y_kmeans == 1, 1], s = 70, c = 'blue', label = 'Cluster 2')
plt.scatter(X_sc[y_kmeans == 2, 0], X_sc[y_kmeans == 2, 1], s = 70, c = 'green', label = 'Cluster 3')
plt.scatter(X_sc[y_kmeans == 3, 0], X_sc[y_kmeans == 3, 1], s = 70, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income($)')
plt.ylabel('Spending Score')
plt.legend(loc = 'upper left')
plt.show()
```
