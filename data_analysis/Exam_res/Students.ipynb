```python
import pandas as pd
```

```python
data = pd.read_csv("data/Students_info.csv").to_dict(orient="records")
```

```python
df = pd.DataFrame(data)
df.index = range(1, len(df) + 1)
```

```python
df.loc[len(df) + 1] = ["Minasyan", "Karen", 20, "Yerevan", 89, 90, 94, 94, 99]
df
```

```python
df["Avg Grade"] = df[[
    "Probability & Statistics",
    "Physics",
    "ComputerScience",
    "DataScience",
    "MachineLearning"
]].mean(axis=1).round(2)
```

```python
df["Result"] = df["Avg Grade"].apply(lambda x: "Passed" if x >= 85 else "Failed")
```

```python
failes = df.loc[df["Result"] == "Failed", "First"]
failes.index = range(1, len(failes) + 1)
failes
```

```python
df.sort_values(by=["Avg Grade", "Last", "First"], ascending=[False, True, True], inplace=True)
df.index = range(1, len(df) + 1)
```

```python
df.First = df.First.str.capitalize()
df.Last = df.Last.str.capitalize()
df.City = df.City.str.capitalize()
```

```python
city = df.groupby("City")
city[["Probability & Statistics", "Physics", "ComputerScience", "DataScience", "MachineLearning"]].agg(["count", "mean", "max", "min", "sum"])
```

```python
df["Category"] = pd.cut(
    df["Avg Grade"],
    bins=[0, 60, 70, 80, 90, 100],
    labels=["F", "D", "C", "B", "A"]
)
df
```

```python
df.to_excel("data/students.xlsx")
```
