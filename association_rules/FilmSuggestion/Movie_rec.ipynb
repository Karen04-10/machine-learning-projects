```python
df = pd.read_csv('data/movielens.csv')
df
```

```python
X = df.drop(columns=['release date', 'unknown', 'movie id'])
```

```python
X
```

```python
transactions = []
for i in range(0, 1682):
    transactions.append([str(X.columns[j]) for j in range(0, 19) if X.values[i, j] == 1])
transactions
```

```python
from apyori import apriori
rules = apriori(transactions=transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=3, max_length=3)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    first = [tuple(result[2][0][0])[0] for result in results]
    second = [tuple(result[2][0][0])[1] if len(tuple(result[2][0][0])) > 1 else '-' for result in results]
    third = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(first, second, third, supports, confidences, lifts))

resultdf = pd.DataFrame(inspect(results), columns=['1st genre', '2nd genre', '3rd genre', 'Support', 'Confidence', 'Lift'])
resultdf
```

```python
resultdf.nlargest(n=30, columns='Lift')
```

```python
movies = X
genre_cols = X.columns[1:]

def movies_with_genre(genre):
    return movies[movies[genre] == 1]['movie title'].tolist()

def recommend_movies(movie_title, top_n=10):
    movie_genres = movies[movies['movie title'] == movie_title][genre_cols]
    if movie_genres.empty:
        return f"{movie_title} not found in dataset."

    movie_genres = movie_genres.values[0]

    suggested_genres = set()
    for index, row in resultdf.iterrows():
        rule_lhs = [row['1st genre'], row['2nd genre']]
        rule_rhs = row['3rd genre']

        for idx, g in enumerate(rule_lhs):
            if g != '-' and movie_genres[genre_cols.get_loc(g)] == 1:
                suggested_genres.add(rule_rhs)

    suggested_movies = []
    for genre in suggested_genres:
        suggested_movies.extend(movies_with_genre(genre))

    suggested_movies = list(set(suggested_movies) - {movie_title})
    return suggested_movies[:top_n]

recommend_movies('Fifth Element, The (1997)', top_n=20)
```
